```python
import os
from google.colab import userdata

# Khởi tạo danh sách các biến cần nạp
keys = ["GEMINI_API_KEY", "USE_MOCK_ASR", "WHISPER_MODEL_NAME", "LLM_MODEL_NAME"]

print("=== NẠP CẤU HÌNH TỪ COLAB SECRETS ===")
for key in keys:
    try:
        val = userdata.get(key)
        if val is not None:
            os.environ[key] = str(val)
            print(f"[+] Đã nạp thành công: {key}")
    except Exception as e:
        print(f"[-] Bỏ qua hoặc lỗi khi nạp {key}: {str(e)}")
```
